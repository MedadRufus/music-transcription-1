```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
from scipy.signal import spectrogram
import soundfile
import os
```

# Spectrogram of a single note

```python
local_path = r'data\IDMT-SMT-GUITAR_V2\dataset1\Fender Strat Clean Neck SC\audio\G53-50110-1111-00011.wav'
abs_path = os.path.join(os.path.dirname(os.path.abspath('')), local_path)
print(abs_path)
samples, sample_rate = soundfile.read(abs_path)
print(samples.shape)
print(sample_rate)
```

```python
f, t, Sxx = spectrogram(samples[4410:16000], sample_rate)
print(f.shape)
print(t.shape)
print(Sxx.shape)
plt.figure(figsize=(10,10))
plt.pcolormesh(t, f, Sxx, cmap='binary')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.ylim(ymax=4000)
```
